Import the functions

```python
import nibabel as nib
import numpy as np
import pandas as pd
import seaborn as sns
import os
from scipy import stats
import matplotlib.pyplot as plt
```

Load data generated by `voxelwise_analysis.ipynb`

```python
df_seed_all = pd.read_pickle('df_seed_all_2_1_23.pkl')
```

Visualize the DMN smoothed on each subject

```python
import matplotlib.colors as colors

cmap_txt = np.loadtxt('/Users/allucas/Documents/research/CNT/P18_7T_rsfMRI/source_data/fsleyes_colormaps/flow.cmap')
flow_cmap = colors.ListedColormap(cmap_txt, 'flow')
flow_cmap
```

```python
mprage_image = nib.load('/Users/allucas/Documents/research/CNT/P18_7T_rsfMRI/source_data/sample_mp2rage/masked_mp2rage.nii.gz').get_fdata()

def plot_4views(image, fname):
    plt.figure(figsize=(20,5))
    plt.subplot(141)
    plt.imshow(np.rot90(image[49,:,:]),vmin=-1, vmax=1, cmap=flow_cmap)
    plt.axis('off')

    plt.subplot(142)
    plt.imshow(np.fliplr(np.rot90(image[:,47,:])),vmin=-1, vmax=1, cmap=flow_cmap)
    plt.axis('off')

    plt.subplot(143)
    plt.imshow(np.fliplr(np.rot90(image[:,:,58])),vmin=-1, vmax=1, cmap=flow_cmap)
    plt.axis('off')

    plt.subplot(144)
    plt.imshow(np.fliplr(np.rot90(image[47,:,:])),vmin=-1, vmax=1, cmap=flow_cmap)
    plt.axis('off')
    plt.tight_layout()
    #plt.savefig('/Users/allucas/Documents/research/CNT/P18_7T_rsfMRI/output/figures/paper_figures/cohensd_maps_all/'+fname)

def plot_4views_smoothed(image, fname):
    plt.figure(figsize=(20,5))
    plt.subplot(141)
    plt.imshow(np.rot90(mprage_image[49,:,:]), cmap='gray')
    plt.imshow(np.rot90(image[49,:,:]),vmin=-0.2, vmax=0.2, cmap=flow_cmap)
    plt.axis('off')

    plt.subplot(142)
    plt.imshow(np.fliplr(np.rot90(mprage_image[:,47,:])),cmap='gray')
    plt.imshow(np.fliplr(np.rot90(image[:,47,:])),vmin=-0.2, vmax=0.2, cmap=flow_cmap)
    plt.axis('off')

    plt.subplot(143)
    plt.imshow(np.fliplr(np.rot90(mprage_image[:,:,58])),cmap='gray')
    plt.imshow(np.fliplr(np.rot90(image[:,:,58])),vmin=-0.2, vmax=0.2, cmap=flow_cmap)
    plt.axis('off')

    plt.subplot(144)
    plt.imshow(np.fliplr(np.rot90(mprage_image[47,:,:])),cmap='gray')
    plt.imshow(np.fliplr(np.rot90(image[47,:,:])),vmin=-0.2, vmax=0.2, cmap=flow_cmap)
    plt.axis('off')
    plt.tight_layout()
    #plt.colorbar()
    #plt.savefig('/Users/allucas/Documents/research/CNT/P18_7T_rsfMRI/output/figures/paper_figures/cohensd_maps_smoothed/'+fname)
    
    
def plot_4views_mprage(image, fname):
    image[image==0] = np.nan
    plt.figure(figsize=(20,5))
    plt.subplot(141)
    plt.imshow(np.rot90(image[49,:,:]), cmap='gray')
    plt.axis('off')

    plt.subplot(142)
    plt.imshow(np.fliplr(np.rot90(image[:,47,:])), cmap='gray')
    plt.axis('off')

    plt.subplot(143)
    plt.imshow(np.fliplr(np.rot90(image[:,:,58])), cmap='gray')
    plt.axis('off')

    plt.subplot(144)
    plt.imshow(np.fliplr(np.rot90(image[47,:,:])), cmap='gray')
    plt.axis('off')
    plt.tight_layout()
    #plt.savefig('/Users/allucas/Documents/research/CNT/P18_7T_rsfMRI/output/figures/paper_figures/cohensd_maps_all/'+fname)

```

```python
source_7t = '/Users/allucas/Documents/research/CNT/P18_7T_rsfMRI/source_data/seed_based_analysis'

for i in df_seed_all[df_seed_all['Lateralization']=='Left']['Record ID'].values:
    dmn_mask_path = os.path.join(source_7t,i,'HarvardOxford/seed_map_difference_dmn_HarvardOxford_smoothed.nii.gz')
    if os.path.exists(dmn_mask_path):
        print(i)
        dmn_mask_diff = nib.load(dmn_mask_path).get_fdata()
        plot_4views_smoothed(dmn_mask_diff,'test')
```

```python
for i in df_seed_all[df_seed_all['Lateralization']=='Right']['Record ID'].values:
    dmn_mask_path = os.path.join(source_7t,i,'HarvardOxford/seed_map_difference_dmn_HarvardOxford_smoothed.nii.gz')
    if os.path.exists(dmn_mask_path):
        dmn_mask_diff = nib.load(dmn_mask_path).get_fdata()
        print(i)
        plot_4views_smoothed(dmn_mask_diff,'test')
```
